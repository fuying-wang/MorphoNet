```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Define the data
strategies = ["Baseline (PatchGCN)", "PANTHER (All)", "OT (All)", "Mean"]
metrics = [
    "EBRAINS (Bal. acc.)", "EBRAINS (F1)", "PANDA KRLS (Cohen's $\kappa$)", "PANDA RUMC (Cohen's $\kappa$)", 
    "BRCA (C-Index)", "KIRC (C-Index)", "LUAD (C-Index)"
]
data = np.array([
    [0.676, 0.742, 0.942, 0.918, 0.694, 0.732, 0.634],
    [0.653, 0.730, 0.903, 0.906, 0.524, 0.653, 0.598],
    [0.654, 0.723, 0.903, 0.899, 0.572, 0.666, 0.609],
    [0.707, 0.768, 0.891, 0.903, 0.745, 0.738, 0.621],
])

indices = np.argsort(data.mean(axis=0))
metrics = np.array(metrics)[indices]
data = data[:, indices]

fig, ax = plt.subplots(1, 1, figsize=(9, 6))
ax.plot(metrics, data[0], label=strategies[0], marker='o', markersize=12, color='#2173A8', linewidth=3)
ax.plot(metrics, data[1], label=strategies[1], marker='^', markersize=12, color='#2F9C79', linewidth=3)
ax.plot(metrics, data[2], label=strategies[2], marker='s', markersize=12, color='#CD5E2B', linewidth=3)
ax.plot(metrics, data[3], label=strategies[3], marker='x', markersize=12, color='#BE76A0', linewidth=3)
# for i in range(len(strategies)):
#     ax.plot(metrics, data[i], label=strategies[i], marker='o')

ax.set_xticklabels(metrics, rotation=45, ha='right', fontsize=13, fontname="Arial")
ax.set_yticklabels([f"{x:.2f}" for x in np.linspace(0.5, 1.0, 11)], fontsize=13, fontname="Arial")

ax.legend(fontsize=12)
ax.grid(linestyle='--')    
ax.set_title("Effect of Aggregation Strategies Across Different Tasks", fontsize=14, fontname="Arial")
plt.savefig("aggregation_strategies.pdf", bbox_inches='tight')
# ax.set_title("Effect of Aggregation Strategies Across Different Datasets", fontsize=16)
```
