```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset and corresponding values
datasets = ['EBRAINS', 'TCGA-NSCLC', 'PANDA', 'TCGA-BRCA', 'TCGA-KIRC', 'TCGA-LUAD']
wsi_nodes = [np.nan, 35174, np.nan, np.nan, np.nan, np.nan]
wsi_edges = [np.nan, 268476, np.nan, np.nan, np.nan, np.nan]
superpatch_nodes = [np.nan, 17, np.nan, np.nan, np.nan, np.nan]
superpatch_edges = [np.nan, 52, np.nan, np.nan, np.nan, np.nan]

x = np.arange(len(datasets))  # Label locations
width = 0.35  # Width of the bars

fig, ax1 = plt.subplots(figsize=(12, 6))

# Use a colormap
colors = plt.get_cmap('viridis')(np.linspace(0.2, 0.8, 2))

# Plot nodes on primary y-axis
ax1.bar(x - width/2, wsi_nodes, width, label='WSI Nodes', color=colors[0])
ax1.bar(x + width/2, superpatch_nodes, width, label='Superpatch Nodes', color=colors[1])
ax1.set_ylabel('Nodes Count (Log Scale)')
ax1.set_yscale('log')
ax1.set_xlabel('Datasets')
ax1.set_xticks(x)
ax1.set_xticklabels(datasets)
ax1.legend(loc='upper left')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Create a secondary y-axis for edges
ax2 = ax1.twinx()
ax2.plot(x, wsi_edges, marker='o', color='tab:blue', label='WSI Edges')
ax2.plot(x, superpatch_edges, marker='o', color='tab:orange', label='Superpatch Edges')
ax2.set_ylabel('Edges Count (Log Scale)')
ax2.set_yscale('log')
ax2.legend(loc='upper right')

plt.title('Graph Scale Comparison with Twin Y-Axes')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data preparation
data = {
    'Dataset': ['EBRAINS (2313)', 'TCGA-NSCLC (1043)', 'PANDA (9549)', 'TCGA-BRCA (1141)', 'TCGA-KIRC (517)', 'TCGA-LUAD (519)', 'Avg.'],
    'WSI Nodes': [10838, 12261, 145, 10999, 14101, 12896, 10206.7],
    'WSI Edges': [81252, 94347, 816, 82947, 108751, 99175, 77881.3],
    'Superpatch Nodes': [18, 14, 6, 15, 15, 14, 13.7],
    'Superpatch Edges': [57, 42, 21, 46, 46, 41, 42.2]
}

dataset_size = np.array([2313, 1043, 9549, 1141, 517, 519])
patient_size = np.array([2141, 946, 9549, 1041, 456, 511])

# Create DataFrame
df = pd.DataFrame(data)
df.drop(index=6, inplace=True)  # Drop the average row

# Calculate ratios
nodes_ratio = df['WSI Nodes'] / df['Superpatch Nodes']
edges_ratio = df['WSI Edges'] / df['Superpatch Edges']
avg_node_ratio = np.mean(nodes_ratio)
avg_edge_ratio = np.mean(edges_ratio)

# Sort by nodes ratio
indices = np.argsort(nodes_ratio)
nodes_ratio = nodes_ratio.values[indices]
edges_ratio = edges_ratio.values[indices]
dataset_size = dataset_size[indices]
patient_size = patient_size[indices]
dataset_names = df['Dataset'].values[indices]
wsi_nodes = df['WSI Nodes'].values[indices]
wsi_edges = df['WSI Edges'].values[indices]

x = np.arange(len(df['Dataset']))
width = 0.3

fig, ax = plt.subplots(figsize=(10, 7))

# Bar plot for ratios
ax.bar(x - width/2, nodes_ratio, width, label='Node Ratio (WSI / Superpatch)', color='#89C3E5', edgecolor='black')
ax.bar(x + width/2, edges_ratio, width, label='Edge Ratio (WSI / Superpatch)', color='#BFAFCF', edgecolor='black')

# Annotate ratio values above bars
for i in range(len(x)):
    ax.text(x[i] - width * 0.5, nodes_ratio[i] * 1.05, f'{int(nodes_ratio[i])}', ha='center', va='bottom', fontsize=10)
    ax.text(x[i] + width * 0.5, edges_ratio[i] * 1.05, f'{int(edges_ratio[i])}', ha='center', va='bottom', fontsize=10)

# Primary axis settings
ax.set_ylabel('WSI / Superpatch Ratio (Log Scale)', fontsize=13)
ax.set_xticks(x)
ax.set_xticklabels(dataset_names, rotation=45, ha='right', fontsize=12)
ax.set_yscale('log')
ax.tick_params(axis='y', labelsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Average ratio lines
ax.axhline(avg_node_ratio, color='#89C3E5', linestyle='--', label='Avg. Node Ratio')
ax.axhline(avg_edge_ratio, color='#BFAFCF', linestyle='--', label='Avg. Edge Ratio')

# Secondary axis for dataset size and patient size
ax2 = ax.twinx()
ax2.plot(x, wsi_edges, marker='o', markersize=8, color='#4E8542', linestyle='-', label='WSI Graph Edges')
ax2.plot(x, wsi_nodes, marker='s', markersize=8, color='#DA534F', linestyle='-', label='WSI Graph Nodes')
# ax2.plot(x, dataset_size, marker='^', markersize=8, color='#1F77B4', linestyle='--', linewidth=2, label='Dataset Size')
# ax2.plot(x, patient_size, marker='D', markersize=8, color='#E5A3C4', linestyle='--', linewidth=2, label='Patient Size')
ax2.set_ylabel('Size (Counts)', fontsize=13)
ax2.tick_params(axis='y', labelsize=12)
ax2.set_ylim(0, 150000)

# Legends
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=3, frameon=False, fontsize=12)

# Save the figure with high resolution and tight layout
plt.tight_layout()
plt.savefig('dataset_size.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
